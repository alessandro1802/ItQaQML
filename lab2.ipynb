```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute, Aer
from qiskit.visualization import *
# Selection of quantum simulator (or processor)
backend = Aer.get_backend('qasm_simulator')
```

## Task 1: Create the Bell state |Î¦+>

```python
qreg_q = QuantumRegister(2, 'q')
creg_c = ClassicalRegister(2, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

circuit.h(qreg_q[0])
circuit.cx(qreg_q[0], qreg_q[1])
```

```python
circuit.measure(qreg_q[0], creg_c[0])
circuit.measure(qreg_q[1], creg_c[1])
circuit.draw(output='mpl', style="clifford")
```

```python
seria = 2048  # Number of trials in a series of measurements of a given type

# Performing quantum calculations
job_simX = execute(circuit, backend, shots=seria)
sim_resultX = job_simX.result()
# Numerical presentation of measurement results
counts = sim_resultX.get_counts(circuit)
print(counts)
```

```python
plot_histogram(counts)
```

```python
plot_distribution(counts)
```
