# Analysis of the operation of quantum gates and simple quantum circuits

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
```

## Task 1
Z-type projection measurement – reading of the qubit state

```python
qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)
```

```python
circuit.measure(qreg_q[0], creg_c[0])
circuit.draw(output='mpl', style={'backgroundcolor': '#EEEEEE'})
```

## Task 2
Quantum gate X = quantum negation

```python
qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)
```

```python
circuit.x(qreg_q[0])
circuit.measure(qreg_q[0], creg_c[0])
circuit.draw(output='mpl', style={'backgroundcolor': '#EEEEEE'})
```

## Task 3: Superposition of states
Operation and reading of the result of Hadamard quantum gate (H gate)

```python
qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)
```

```python
circuit.h(qreg_q[0])
circuit.measure(qreg_q[0], creg_c[0])
circuit.draw(output='mpl', style={'backgroundcolor': '#EEEEEE'})
```

## Task 4: State tomography of one qubit – implementations on a quantum computer

### Measurement in the X base

```python
qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)
```

```python
circuit.ry(pi / 2, qreg_q[0])
circuit.p(pi / 2, qreg_q[0])
circuit.h(qreg_q[0])
circuit.measure(qreg_q[0], creg_c[0])
circuit.draw(output='mpl', style={'backgroundcolor': '#EEEEEE'})
```

### Measurement in the Y base

```python
qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)
```

```python
circuit.ry(pi / 2, qreg_q[0])
circuit.p(pi / 2, qreg_q[0])
circuit.sdg(qreg_q[0])
circuit.h(qreg_q[0])
circuit.measure(qreg_q[0], creg_c[0])
circuit.draw(output='mpl', style={'backgroundcolor': '#EEEEEE'})
```

### Measurement in the Z base

```python
qreg_q = QuantumRegister(4, 'q')
creg_c = ClassicalRegister(4, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)
```

```python
circuit.ry(pi / 2, qreg_q[0])
circuit.p(pi / 2, qreg_q[0])
circuit.measure(qreg_q[0], creg_c[0])
circuit.draw(output='mpl', style={'backgroundcolor': '#EEEEEE'})
```
